```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('./merged_matched_full.csv')
df.head()
```

```python
def describe_col(col):
    print(df[col].describe())
    print(df[col].value_counts(dropna=False))
    print(df[col].hist())
    print(df[col].isnull().sum(axis = 0))
```

```python
def fill_median(col):
    median = df[col].median()
    df[col] = df[col].fillna(median)
```

```python
fill_median('elnantr')
fill_median('iderncvr')
```

```python
df['trn/dep'] = df['trn'] / df['dep'] # Transaction accounts/Total deposits
df['elnantr/elnatqr'] = df['elnantr'] # Credit loss provision to net charge-offs; No elnatqr
df['intexpy/intexpyq'] = df['intexpy'] # Cost of funding earning assets; No intexpyq
df['noniiay/noniiayq'] = df['noniiay'] # Noninterest income to average assets; No noniiayq
df['nonixay/nonixayq'] = df['nonixay'] # Noninterest expense to average assets; No nonixayq
df['iderncvr/iderncvq'] = df['iderncvr'] # Earnings coverage of net charge-offs (x); No iderncvq
df['ernastr/ernastr'] = df['ernastr'] # Doesn't make sense when ratio is 1; # Earning assets to total assets ratio \
#df['ernastr'] / df['ernastr'] # Earning assets to total assets ratio
df['nperfv/nperfv'] = df['nperfv'] # Noncurrent assets plus other real estate owned to assets
```
